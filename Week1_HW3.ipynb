```python
import math

loanAmount = float(input("Loan Amount: "))
numberOfYears  = int(input("Number of Years: "))
annualInterestRate = float(input("Annual Interest Rate: "))

monthlyInterestRate = annualInterestRate / 1200

monthlyPayment = loanAmount * monthlyInterestRate / (1 - 1 / pow(1 + monthlyInterestRate, numberOfYears * 12))

print("\nMonthly Payment: " + str(round(monthlyPayment, 2)))
print("Total Payment: " + str(round(monthlyPayment * 12 * numberOfYears, 2)))

balance = loanAmount

print("\nPayment#\t Interest\t Principal\t\t Balance")
for i in range(1, numberOfYears * 12 + 1):
    interest = monthlyInterestRate * balance
    principal = monthlyPayment - interest
    balance = balance - principal
    print(i, "\t\t", '{:.{}f}'.format(interest, 2), "\t\t", '{:.{}f}'.format(principal, 2), "\t\t\t",
        '{:.{}f}'.format(balance, 2))
```
