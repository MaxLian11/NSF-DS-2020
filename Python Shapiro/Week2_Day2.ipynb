```python
s = input("Enter scores: ")
items = s.split()
student = [eval(x) for x in items]
sorted_student = list(student)

def grade(score):
    if score >= max(student) - 10:
        return 'A'
    elif score >= max(student) - 20:
        return 'B'
    elif score >= max(student) - 30:
        return 'C'
    elif score >= max(student) - 40:
        return 'D'
    else:
        return 'F'

for x in range(len(student)):
    print("Student " + str(x) + " is " + str(student[x]) + " and grade is " + grade(student[x]))
```

```python
s = input("Enter integers between 1 and 100: ")
items = s.split()
arr = [eval(x) for x in items]

print_list = [] # store strings in a list

for i in range(len(arr)):
    if arr[i] not in arr[:i]: # don't print 
        print_list.append(str(arr[i]) + " occurs " + str(arr.count(arr[i])))
        # print "time" or "times" depending on the number of occurances
        if arr.count(arr[i]) == 1: print_list[-1] += " time"
        else: print_list[-1] += " times"
# print a sortd list depenting on first 3 characters of a string (number from from "1" to "100")
print(*sorted(print_list, key=lambda x: int(x.split(' ')[0])), sep='\n')
```

```python
matrix = [[],[],[]]

def sumColumn(m, columnIndex):
    total = 0
    for i in range(len(m)):
        total += m[i][columnIndex]
    return total

for i in range(len(matrix)):
    s = map(float, input("Enter a 3-by-4 matrix row for row " + str(i) + ": ").split())
    matrix[i] = [x for x in s]

for i in range(len(matrix[0])):
    print("Sum of the elements for column " + str(i) + " is " + str(sumColumn(matrix, i)))
    # print(sum(row[i] for row in matrix))
    
```

```python

matrix = [[],[],[],[]]
def sumMajorDiagonal(m):
    total = 0
    for i in range(len(m)):
        total += m[i][i]
    return total
for i in range(len(matrix)):
    s = map(float, input("Enter a 4-by-4 matrix row for row " + str(i+1) + ": ").split())
    matrix[i] = [x for x in s]

print("Sum of the elements in the major diagonal is " + str(sumMajorDiagonal(matrix)))
```
