```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
dataset = pd.read_csv("AnscombesQuartet.csv")
x1 = np.array(dataset["x1"])
y1 = np.array(dataset["y1"])
x2 = np.array(dataset["x2"])
y2 = np.array(dataset["y2"])
x3 = np.array(dataset["x3"])
y3 = np.array(dataset["y3"])
x4 = np.array(dataset["x4"])
y4 = np.array(dataset["y4"])
```

```python
fig,ax=plt.subplots(ncols=2,nrows=2)
ax[0][0].scatter(x1,y1)
ax[0][1].scatter(x2,y2)
ax[1][0].scatter(x3,y3)
ax[1][1].scatter(x4,y4)

plt.gcf().set_size_inches(16,7)
```

```python
print(linregress(x1,y1), "\n")
print(linregress(x2,y2), "\n")
print(linregress(x3,y3), "\n")
print(linregress(x4,y4), "\n")
```

```python
fig,ax=plt.subplots(ncols=2,nrows=2)

x_list = [x1,x2,x3,x4]
y_list = [y1,y2,y3,y4]
ax_list = [ax[0][0],ax[0][1],ax[1][0],ax[1][1]]

for i in range(4):
    slope, intercept, r_value, p_value, std_err = linregress(x_list[i],y_list[i])
    ax_list[i].scatter(x_list[i],y_list[i])
    ax_list[i].plot(x_list[i], intercept + slope*x_list[i], 'r', label=round(slope,5))
    ax_list[i].legend(loc = 'upper left')

plt.gcf().set_size_inches(16,10)
```
